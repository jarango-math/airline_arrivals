```python
df=pd.read_csv('2000.csv.bz2')
```

```python
df.isnull().sum()
```

```python
df=df.drop(['Year','Month','DayofMonth','DayOfWeek','CancellationCode','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay'],axis=1)
```

```python
df['late']=np.where((df['ActualElapsedTime']-df['CRSElapsedTime'])>=30,1,0)
```

```python
df=df.drop('TailNum',axis=1)
```

```python
df=df.drop(['UniqueCarrier','FlightNum'],axis=1)
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import cross_val_score
```

```python
bnb = BernoulliNB()
X = df.drop('late', 1)
Y = df['late']
X = pd.get_dummies(X)
cross_val_score(bnb, X, Y, cv=3)
```
